```python
report_card_dictionary = {"student_names" : ["zaid", "jay", "aryan", "sahil", "riya"],
                          "math_scores" : [50, 43,48,45,31],
                          "chemistry_scores" : [23, 43,68,85,11],
                          "information_technology_scores" : [51,33,58,25,81],
                          "physics_scores" : [60, 73,28,90,21]}
report_card_dictionary
```

```python
#setup
import pandas as pd
```

```python
#converting dictionary to dataframe
report_card_df = pd.DataFrame(report_card_dictionary)
report_card_df
```

```python
#showing only 10 rows using head function
report_card_df.head()
```

```python
report_card_df.head(2)
```

```python
#sample of 2 rows in our data frame
report_card_df.sample(2)
```

```python
report_card_df.describe()
```

```python
#create a new column called total score
report_card_df["total_score"] = report_card_df["math_scores"] + report_card_df["chemistry_scores"] + report_card_df["information_technology_scores"] + report_card_df["physics_scores"]
report_card_df.head()
```

```python
#create a colmun called average score
report_card_df["average_score"] = report_card_df["total_score"]/4
report_card_df.head()
```

```python
report_card_df.describe()
```

INSIGHTS
what is the mean chemistry score = 46

```python
# CREATING A TABLE CALLED REPORT CARD INSIGHTS
report_card_insights_df = report_card_df.describe().transpose()
report_card_insights_df
```

```python
#creating a csv file of our report card insights

report_card_insights_df.to_csv("report card insights.csv")
```
